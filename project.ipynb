```python
from ipyleaflet import Map, basemaps, Marker

# Map centred on (60 degrees latitude et -2.2 degrees longitude)
# Latitude, longitude
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.CartoDB.Positron)
map
```

```python
## HERE: plot the coordinates of the estates
#Extrayendo coordenadas y poblacion del subset
coordinates= fil_ds[['latitude','longitude']].values
populations=fil_ds['level5'].values

## PUT HERE YOUR CODE:
#agregando marcadores para el mapa con popup
for(lat, lon), pop in zip(coordinates,populations):
    marker=Marker(location=(lat,lon))
    map.add_layer(marker)

map
```
